# Task :  
What is the Annualized  Return of the person who bids in the last month ?

What is the Annualized Return of the person who bids in the first month ?

Write an R/Python script which calculates the annualized return of chit fund participant ?

Show the Return % for each month's bid winner. 

```python
import pandas as pd
```

# Reading provided datafile

```python
d = pd.DataFrame(pd.read_excel("ChitFundData.xlsx", skiprows=1))
```

```python
d.head()
```

# Python Script

```python
for i in d.Month:
    print("Month : \t\t",i)
    bid_ret = d['Net amount recd by Bid winner'][i-1]
    #print("won bid : \t\t",bid_ret)    

    month_ret = d['Amount returned to everyone in the group'][i-1]
    #print("amount return\t\t",month_ret)
    
    for a in d['Amount returned to everyone in the group']:
        bid_ret = bid_ret+a
    print("Total Return \t\t",bid_ret)    
    
    annual_ret = bid_ret*(12/25)
    print("Annual Return \t\t",annual_ret)
    
    ret_percnt = bid_ret*(100/50000)
    print("Return percent \t\t",ret_percnt)
    
    print("")
    
```

# Answers

Annualized  Return of the person who bids in the last month: 25776.0

Annualized Return of the person who bids in the first month: 20976.0
